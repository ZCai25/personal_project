```python
import pandas as pd

# read the json file as a pandas dataframe
df = pd.read_json('/content/tables.json') 

# display the dataframe
df.shape

df['entries'][0]

list_of_dicts = df['entries'].to_list()
b = list_of_dicts[0]

for dictionary in b:
    print(dictionary["rowNum"] + ',' + dictionary['tableItem'])
```

```python
df.columns
```

```python
def extract_info(entries):
    result = ''
    for dictionary in entries:
        result += f"{dictionary['rowNum']},{dictionary['tableItem']};"
    return result

# Apply the function to the 'entries' column using the apply method
df['extracted_info'] = df['entries'].apply(extract_info)


df_new = df[['name' ,'extracted_info']].copy()
df_new.head()

df_new['extracted_info'] = df_new['extracted_info'].str.replace('\*\*', '')

df_new.head()

```

```python
df_new.shape
```

```python
df_new_v1 = df_new.copy()
df_new_v1['concatenated_column'] = df_new['name'] + df_new['extracted_info']
df_new_v1
```

```python
df_new_v1.to_csv('/content/prompts.csv')
```
