```python
import requests
from bs4 import BeautifulSoup
import csv
import praw
import re
from user_definition import *
import pickle
```

```python
def praw_setup(client_id, client_secret, user_agent, password, username):
    '''
    Instantiate the Python Reddit API Wrapper (PRAW)
    object in order to access Reddit data.
    
    client_id = client_id from your app info on 
        Reddit's dev website
    client_secret = client_secret from from your
        app info on Reddit's dev website
    user_agent = A string representing whoever is
        accessing the data. Per Reddit's API rules,
        must include your Reddit username.
    password = Your reddit account's password.
    username = Your reddit username.
    '''
    reddit = praw.Reddit(
        client_id=client_id,
        client_secret=client_secret,
        user_agent=user_agent,
        password=password,
        username=username
    )
    return reddit
```

```python
# Authenticate with the Reddit API
reddit = praw_setup(client_id, client_secret, user_agent, password, username)

# Get the subreddit and the post
subreddit = reddit.subreddit("d100")
post = subreddit.wiki["finished"]

# Extract all the links in the post
links = []
for paragraph in post.content_md.split("\n\n"):
    str_ = re.findall(r'\[(.*?)\]\((.*?)\)', paragraph)
    for word in str_:
        links.append(word[1])

# Loop through each link and check if it contains a table
for link in links:
    # print(link)
    response = requests.get(link)
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the table element using its class name
    table = soup.find(class_="post-content-wrap")
    # print(table)
    # If a table is found, extract its headers and data rows and save them to a CSV file
    if table is not None:
        # Extract the table headers and data rows
        headers = [header.text for header in table.find_all("th")]
        data_rows = []
        for row in table.find_all("tr"):
            data_rows.append([data.text for data in row.find_all("td")])
        # Save the table to a CSV file
        filename = link.split("/")[-2] + ".csv"
        # print(filename)
        with open(f"d100/{filename}", "w", newline="") as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(headers)
            writer.writerows(data_rows)
            print(f"{filename} saved successfully.")
        
```

```python
# put all the csv files in one text file
import glob

# List all CSV files in the folder
csv_files = glob.glob('scrape_result_d100/*.csv')

# Open a new text file to write the combined data
with open('scrape_result_txt/combined_data_d100.txt', 'w') as f:

    # Loop through each CSV file
    for file in csv_files:
        
        # Open the file and read the data
        with open(file, 'r') as csv_file:
            data = csv_file.read()
            
        # Write the data to the text file with a newline character at the end
        f.write(data)
        f.write('\n')
 
```
